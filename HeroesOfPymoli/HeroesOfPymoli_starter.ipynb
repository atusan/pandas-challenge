```python
# Dependencies and Setup
import pandas as pd
from scipy.stats import mode
from IPython.display import display
```

```python

# File to Load (Remember to Change These)
file_to_load = "Resources/purchase_data.csv"

# Read Purchasing File and store into Pandas data frame
purchase_data = pd.read_csv(file_to_load)
purchase_data.head()
```

## Player Count

* Display the total number of players


```python
Total_Players = purchase_data['SN'].unique()

Total_Players = len(Total_Players)

summary_table = pd.DataFrame([{'Total Players': Total_Players}])
summary_table
```

## Purchasing Analysis (Total)

* Run basic calculations to obtain number of unique items, average price, etc.


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame


```python
num_uni_item = len(purchase_data['Item ID'].unique())

avg_price = round(purchase_data['Price'].mean(),2)

num_purchases = purchase_data['Purchase ID'].count()

total_rev = round(purchase_data['Price'].mean()*num_purchases,2)

summary_table = pd.DataFrame([{'Number of Unique Items': num_uni_item,'Average Price':'$'+ str(avg_price),
                               'Number of Purchases':num_purchases,
                               'Total Revenue':'$'+ str(total_rev)}])
summary_table
```

## Gender Demographics

* Percentage and Count of Male Players


* Percentage and Count of Female Players


* Percentage and Count of Other / Non-Disclosed




```python

total_count_gender= purchase_data.groupby('Gender').nunique()['SN']

percentage_count = round((total_count_gender/Total_Players)*100,2)


summary_table = pd.DataFrame({"Total Count": total_count_gender,
                              "Percentage of Players": percentage_count
    
                                   })
summary_table['Percentage of Players'] =  summary_table['Percentage of Players'].map("{:.2f}%".format)                           

summary_table.reset_index(inplace=True)
summary_table
```


## Purchasing Analysis (Gender)

* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. by gender




* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_count = purchase_data.groupby('Gender').nunique()['Purchase ID']
purchase_count

purchase_price_avg = purchase_data.groupby('Gender')['Price'].mean()
purchase_price_avg

purchase_total = purchase_data.groupby('Gender')['Price'].sum()
purchase_total

purchase_total_avg = purchase_data.groupby('Gender')['Price'].sum()/purchase_data.groupby('Gender').nunique()['SN']

summary_table = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": purchase_price_avg,
                              "Total Purchase Value": purchase_total,
                              "Avg Total Purchase per Person": purchase_total_avg
    
                                   })
summary_table['Average Purchase Price'] = summary_table['Average Purchase Price'].map("${:.2f}".format)

summary_table['Total Purchase Value'] = summary_table['Total Purchase Value'].map("${:.2f}".format)

summary_table['Avg Total Purchase per Person'] = summary_table['Avg Total Purchase per Person'].map("${:.2f}".format)


summary_table 
```

## Age Demographics

* Establish bins for ages


* Categorize the existing players using the age bins. Hint: use pd.cut()


* Calculate the numbers and percentages by age group


* Create a summary data frame to hold the results


* Optional: round the percentage column to two decimal points


* Display Age Demographics Table


```python
num_max = purchase_data['Age'].max()
num_max=45
num_min = purchase_data['Age'].min()
num_min=7


bins = [0,9,14,19,24,29,34,39,45]

# Create labels for these bins
group_labels = ["<10", "10-14", "15-19","20-24", "25-29","30-34","35-39","40+"]

pd.cut(purchase_data["Age"], bins, labels=group_labels).head()

purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)
purchase_data["Age Ranges"]

total_count_age_group = purchase_data.groupby("Age Ranges").nunique()['SN']

percentage_players = total_count_age_group /Total_Players*100


summary_table = pd.DataFrame({"Total Count": total_count_age_group,
                              "Percentage of Players": percentage_players
                            
                             })

summary_table['Percentage of Players'] = summary_table['Percentage of Players'].map("{:.2f}%".format)

summary_table 
```

## Purchasing Analysis (Age)

* Bin the purchase_data data frame by age


* Run basic calculations to obtain purchase count, avg. purchase price, avg. purchase total per person etc. in the table below


* Create a summary data frame to hold the results


* Optional: give the displayed data cleaner formatting


* Display the summary data frame

```python
purchase_data["Age Ranges"] = pd.cut(purchase_data["Age"], bins, labels=group_labels)

purchase_count = purchase_data.groupby("Age Ranges").nunique()["Purchase ID"]

total_purchase_price = purchase_data.groupby("Age Ranges")["Price"].sum()

avg_purchase_price = purchase_count/total_purchase_price

avg_total_purchase_person = total_purchase_price/purchase_data.groupby("Age Ranges").nunique()['SN']

summary_table = pd.DataFrame({"Purchase Count": purchase_count,
                              "Average Purchase Price": avg_purchase_price,
                              "Total Purchase Value": total_purchase_price,
                              "Avg Total Purchase per Person": avg_total_purchase_person
    
                                   })
summary_table['Average Purchase Price'] = summary_table['Average Purchase Price'].map("${:.2f}".format)

summary_table['Total Purchase Value'] = summary_table['Total Purchase Value'].map("${:.2f}".format)

summary_table['Avg Total Purchase per Person'] = summary_table['Avg Total Purchase per Person'].map("${:.2f}".format)


summary_table 
```

## Top Spenders

* Run basic calculations to obtain the results in the table below


* Create a summary data frame to hold the results


* Sort the total purchase value column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python

purchase_count_sn = purchase_data.groupby('SN').nunique()["Purchase ID"]



# purchase_count_sn = purchase_data.groupby('SN')
# dfb=purchase_count_sn["Purchase ID"].count()

total_purchase_price = purchase_data.groupby("SN")["Price"].sum()
total_purchase_price

avg_total_purchas_price = purchase_count_sn/total_purchas_price
avg_total_purchas_price

summary_table = pd.DataFrame({"Purchase Count": purchase_count_sn,
                              "Average Purchase Price": avg_total_purchas_price,
                              "Total Purchase Value": total_purchas_price,
                             })

summary_table = summary_table.sort_values('Total Purchase Value', ascending=False)

summary_table['Average Purchase Price'] = summary_table['Average Purchase Price'].map("${:.2f}".format)

summary_table['Total Purchase Value'] = summary_table['Total Purchase Value'].map("${:.2f}".format)

summary_table.head()
```

## Most Popular Items

* Retrieve the Item ID, Item Name, and Item Price columns


* Group by Item ID and Item Name. Perform calculations to obtain purchase count, item price, and total purchase value


* Create a summary data frame to hold the results


* Sort the purchase count column in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the summary data frame



```python
# how I can get the price based on purchsed ID or Item ID directly since 
# I aldready have it so I dont have to get the total price and then divide it by purchase count to get price of each item

item_data = purchase_data[['Item ID','Item Name','Price']]

item_grouped = item_data.groupby(['Item ID','Item Name'])

purchase_count = item_grouped['Price'].count()

total_item_price = item_grouped['Price'].sum()

each_item_price = total_item_price / purchase_count

summary_table = pd.DataFrame({"Purchase Count": purchase_count,
                              "Item Price": each_item_price,
                              "Total Purchase Value": total_item_price,
                             })

summary_table = summary_table.sort_values('Purchase Count', ascending=False)

summary_table['Item Price'] = summary_table['Item Price'].map("${:.2f}".format)

summary_table['Total Purchase Value'] = summary_table['Total Purchase Value'].map("${:.2f}".format)

summary_table.head(20)
```

## Most Profitable Items

* Sort the above table by total purchase value in descending order


* Optional: give the displayed data cleaner formatting


* Display a preview of the data frame



```python
item_data = purchase_data[['Item ID','Item Name','Price']]

item_grouped = item_data.groupby(['Item ID','Item Name'])

purchase_count = item_grouped['Price'].count()

total_item_price = item_grouped['Price'].sum()

each_item_price = total_item_price / purchase_count

summary_table = pd.DataFrame({"Purchase Count": purchase_count,
                              "Item Price": each_item_price,
                              "Total Purchase Value": total_item_price,
                             })

summary_table = summary_table.sort_values('Total Purchase Value', ascending=False)

summary_table['Item Price'] = summary_table['Item Price'].map("${:.2f}".format)

summary_table['Total Purchase Value'] = summary_table['Total Purchase Value'].map("${:.2f}".format)

summary_table.head(20)
```
